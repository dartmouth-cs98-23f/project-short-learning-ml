```python
import os
from pprint import pprint 
from sklearn.preprocessing import LabelEncoder
import ast
import pandas as pd
from clipper import clipper
import ffmpeg
```

```python
data_file = 'AhW3jlg7NQw_clip.csv'
video_file = 'AhW3jlg7NQw.mp4'
data_dir = '../../../app/data/'
output_dir = 'output/'
video_dir = 'videos/'

file_path = data_dir+output_dir+data_file
video_path = data_dir+video_dir+video_file

df = pd.read_csv(file_path)
df.head()
```

```python
# convert the string representation of results to a list
df['list_categories'] = df['categories'].apply(ast.literal_eval)

# get most likely label from each list
df['label'] = df['list_categories'].apply(lambda x: x[0] if x else None)
```

```python
# initialize the LabelEncoder
label_encoder = LabelEncoder()

# fit and transform the label to a numerical value
numerical_data = label_encoder.fit_transform(df['label'])

df['label_value'] = numerical_data
```

```python
topics = list(df['label_value'])
timestamps = list(df['second'])
window=30
threshold = 0.5

result = clipper(topics, timestamps, window, 'mean', threshold)
```

```python
def trim_video(input_file, output_file, start_time, end_time):
  (
    ffmpeg.input(input_file, ss=start_time, to=end_time)
    .output(output_file)
    .run()
  )
```

```python
clips_dir = './clips/'
for start, end in result:
    if end-start < 15:
        continue

    start_time = str(start)
    end_time = str(end)

    output_file = clips_dir+start_time+'-'+end_time+'.mp4'
    print(output_file)
    trim_video(video_path, output_file, start_time, end_time)
```
